```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
pharma_data_train = pd.read_csv("Train_Data.csv")
```

```python
pharma_data_train.head()
```

```python
pharma_data_train.info()
```

```python
pharma_data_train.size
```

```python
pharma_data_train.isnull().sum()
```

```python
pharma_data_train=pharma_data_train.dropna()
pharma_data_train.isnull().sum()
```

```python
# Encode the features to integers
categorical_col=pharma_data_train[["Treated_with_drugs", "Patient_Smoker", "Patient_Rural_Urban", "Patient_mental_condition"]]

encoder=LabelEncoder()

for i in categorical_col.columns:
    pharma_data_train[i]=encoder.fit_transform(pharma_data_train[i])
```

```python
pharma_data_train.head()
```

```python
pharma_data_train.set_index("Patient_ID", inplace=True) #Dataset Id's
pharma_data_train.head()
```

```python
pharma_data_train.corr()['Survived_1_year']
```

```python
# Split the dataset into train and test sets
train, test = train_test_split(pharma_data_train, test_size=0.2, random_state=42)
```

```python
y_train = train.loc[:,"Survived_1_year"]
X_train = train.drop("Survived_1_year",axis=1)

y_test = test.loc[:,"Survived_1_year"]
X_test = test.drop("Survived_1_year",axis=1)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.fit_transform(X_test)
```

```python
# Training models
logistic_model=LogisticRegression(random_state=42)
logistic_model.fit(X_train_sc,y_train)

ridge_model=RidgeClassifier()
ridge_model.fit(X_train,y_train)

knn_model=KNeighborsClassifier(n_neighbors=3, metric="minkowski")
knn_model.fit(X_train,y_train)

svc_model=LinearSVC()
svc_model.fit(X_train_sc,y_train)

decision_tree_model=DecisionTreeClassifier(criterion="entropy")
decision_tree_model.fit(X_train,y_train)

random_forest_model=RandomForestClassifier(n_estimators=100, criterion="entropy")
random_forest_model.fit(X_train,y_train)

naive_bayes_model=GaussianNB()
naive_bayes_model.fit(X_train,y_train)

neural_network_model=MLPClassifier(hidden_layer_sizes=(150,100,50), max_iter = 300,activation = 'relu', solver = 'adam')
neural_network_model.fit(X_train_sc,y_train)
```

```python
# Make predictions
logistic_predictions=logistic_model.predict(X_test_sc)

ridge_predictions=ridge_model.predict(X_test)

knn_predictions = knn_model.predict(X_test)

svc_predictions=svc_model.predict(X_test_sc)

tree_predictions=decision_tree_model.predict(X_test)

random_forest_predictions=random_forest_model.predict(X_test)

naive_bayes_predictions=naive_bayes_model.predict(X_test)

neural_network_predictions=neural_network_model.predict(X_test_sc)
```

```python
# Confusion Matrix
logistic_c_matrix=confusion_matrix(y_test, logistic_predictions)

ridge_c_matrix=confusion_matrix(y_test, ridge_predictions)

knn_c_matrix=confusion_matrix(y_test, knn_predictions)

svc_c_matrix=confusion_matrix(y_test, svc_predictions)

tree_c_matrix=confusion_matrix(y_test, tree_predictions)

random_forest_c_matrix=confusion_matrix(y_test, random_forest_predictions)

naive_bayes_c_matrix=confusion_matrix(y_test, naive_bayes_predictions)

neural_network_c_matrix=confusion_matrix(y_test, neural_network_predictions)
```

```python
# Print reports
print("---------- Logistic Regression ----------")
print(logistic_c_matrix)

print("---------- Ridge Regression ----------")
print(ridge_c_matrix)

print("---------- KNN Classifier ----------")
print(knn_c_matrix)

print("---------- Support Vector Classifier ----------")
print(svc_c_matrix)

print("---------- Decision Tree ----------")
print(tree_c_matrix)

print("---------- Random Forest ----------")
print(random_forest_c_matrix)

print("---------- Naive Bayes ----------")
print(naive_bayes_c_matrix)

print("---------- Neural Network ----------")
print(neural_network_c_matrix)
```

```python
# Classification Reports
logistic_report=classification_report(y_test, logistic_predictions)

ridge_report=classification_report(y_test, ridge_predictions)

knn_report = classification_report(y_test, knn_predictions)

svc_report=classification_report(y_test, svc_predictions)

tree_report=classification_report(y_test, tree_predictions)

random_forest_report=classification_report(y_test, random_forest_predictions)

naive_bayes_report=classification_report(y_test, naive_bayes_predictions)

neural_network_report=classification_report(y_test, neural_network_predictions)
```

```python
# Print reports
print("---------- Logistic Regression ----------")
print(logistic_report)

print("---------- Ridge Regression ----------")
print(ridge_report)

print("---------- KNN Classifier ----------")
print(knn_report)

print("---------- Support Vector Classifier ----------")
print(svc_report)

print("---------- Decision Tree ----------")
print(tree_report)

print("---------- Random Forest ----------")
print(random_forest_report)

print("---------- Naive Bayes ----------")
print(naive_bayes_report)

print("---------- Neural Network ----------")
print(neural_network_report)
```
